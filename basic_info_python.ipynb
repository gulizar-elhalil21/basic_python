```python
#if we put the numbers in the quotation marks they will be considered as strings and when we add them together it will concatenate them instead of adding them mathematically
import math
num1 = "10" #type of the num 1 is string
num2 = "20" #type of the num 2 is string

reslut = num1 + num2
print('the result is: ', reslut)
print(type(reslut))
```

```python
#if we want to add the numbers mathematically we need to convert them to integers or floats
import math
num1 = 10 #type of the num 1 is integer
num2 = 20 #type of the num 2 is interger

reslut = num1 + num2
print('the result is: ', reslut)
print(type(reslut))
```

```python
# if we want to calculatet the area of a circle we can use the formula A = πr^2
import mpmath # first we import the math library to use the value of pi
radius = 3 # we define the radius of the circle
area = mpmath.pi * radius ** 2 # we calculate the area of the circle
print('the area of the circle is: ', area)
```
